```python
import pandas as pd
import numpy as np
```

```python
n = pd.read_csv('/content/Social_Network_Ads.csv')
```

```python
n.head()
```

```python
n.isna().sum()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,r2_score
```

```python
X = n[['Age' , 'EstimatedSalary']]
```

```python
y = n['Purchased']
```

```python
x_t , x_te , y_t ,y_te = train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
x_t
```

```python
x_te
```

```python
y_t
```

```python
y_te
```

```python
model = RandomForestClassifier()
```

```python
model.fit(x_t , y_t)
```

```python
p = model.predict(x_te)
```

```python
p
```

```python
r2_score(p,y_te)
```
